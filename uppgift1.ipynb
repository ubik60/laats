## 3. Energi och skalärprodukt

Man kan införa skalärprodukten 
$$ 
\def\ol#1{\overline{#1}}
\langle f, g \rangle = \int f(t) \ol{g(t)} \, dt
$$
för signaler $f(t)$, $g(t)$ med ändlig energi. Beteckningen $\ol{g(t)}$ betyder komplexkonjugatet av 
signalen $g(t)$. 

**Energin** hos en signal definieras som 
$$
\| f \|^2 = \langle f, f\rangle = \int |f|^2 \, dt.
$$
För diskreta signaler byts integraler ut mot en summa.


**(a)** Beräkna energin $\int_{-\infty}^\infty |x(t)|^2 \, d t$ av signalen i uppgift 2 ovan: (a) med hjälp av `trapz(...)` (b) med hjälp av `sum(..)` och lämplig normalisering.

**(b)** Definiera skalärprodukten $\langle x, y \rangle$ som en funktion `scp(t,x,y)` med tre argument. Det första är tidsvektorn `t` och sedan de två signalerna `x` och `y`.


```octave
trapz(t,x(t).^2) # (a)
```

### Lösning Skalärprodukten för signaler med ändlig energi. (Ej periodiska tidskontinuerliga)

Här definierar vi skalärprodukten som en funktion `scp(t,x,y)` med tre argument. Det första är tidsvektorn och sedan de två signalerna `x` och `y`.
Här beräknas energin av signalen 

```octave
scp = @(t,x,y) trapz(t,x .* conj(y)) 

# Test
t =  linspace(-4,4,1000);
x = @(t) (1-t/3) .* (t<3 & t>0)
scp(t,x(t),x(t))
```
